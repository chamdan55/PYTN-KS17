## Pass by reference vs value

```python
def changeMe(mylist):
    '''
    This changes a passed list into this function
    '''
    mylist.append([1, 2, 3, 4])
    print('values inside the function: ', mylist)
    return
```

```python
mylist = [10, 20, 30]
changeMe(mylist)
print('values inside the function: ', mylist)
```

```python
def change_me(mylist):
    '''
    This changes a passed list into this function
    '''
    mylist = [1, 2, 3, 4]
    print('values inside the function: ', mylist)
    return
```

```python
mylist = [10, 20, 30]
change_me(mylist)
print('values inside the function: ', mylist)
```

## Function Arguments

```python
# Required Arguments
def printme(str):
    '''
    This print a passed string into this function
    '''
    print(str)
    return
```

```python
printme()
```

```python
# Keyword Arguments
def printInfo(name, age):
    '''
    This Print Name and Age
    '''
    print('Name :', name)
    print('Age :', age)
    return
```

```python
printInfo(age=4, name='Hacktiv8')
```

```python
# Default Arguments
def printInfo(name, age=30):
    '''
    This Print Name and Age
    '''
    print('Name :', name)
    print('Age :', age)
    return
```

```python
printInfo('Hacktiv8', 12)
printInfo('Fulan')
```

```python
# Variable-length Arguments
```

```python
def print_info(arg1, *vartuple):
    '''
    this prints a variable passed arguments
    '''
    print('Output is :')
    # print(arg1)
    for val in vartuple:
        print(val)
    return
```

```python
print_info(10)
```

```python
print_info(70, 3.2, 'a', True)
```

## Anonymous Function

```python
sum_1 = lambda arg1, arg2: arg1 + arg2

def sum_2(arg1, arg2):
    return arg1 + arg2
```

```python
print('2 + 3 from sum_1 : ', sum_1(2,3))
print('2 + 3 from sum_2 : ', sum_2(2,3))
```

## Return Statement

```python
def sum(arg1, arg2):
    '''
    Add both paramaters and return them.
    '''
    total = arg1 + arg2
    total2 = total + arg1
    print('Inside the Function :', total)
    return total2
```

```python
total = sum(10, 20)
print('Outside the Function :', total)
```

## Scope of Variable

```python
# Global vs. Local Variable

total = 0

def sum(arg1, arg2):
    total = arg1 + arg2
    print('Inside the function local total :', total)
    return total

sum(10,20)
print('Outside the function global total :', total)
```

```python
jumlah_kucing = 20

def jumlah_hewan():
    jumlah_anjing = 30
    return jumlah_kucing + jumlah_anjing

def jumlah_kucing2():
    return jumlah_kucing + jumlah_kucing

print(jumlah_hewan())
print(jumlah_kucing2())
```
