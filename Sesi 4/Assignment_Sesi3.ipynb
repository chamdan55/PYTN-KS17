```python
def jangkauan(arr):
    '''
    Mengembalikan nilai Jangkauan(Range) dari sebuah array. Input berupa sebuah Array.
    '''
    return arr.max() - arr.min()
```

```python
def quartil(arr):
    import numpy as np
    arr = np.sort(arr)
    
    q3 = np.quantile(arr, .75)
    q1 = np.quantile(arr, .25)
    
    result = (q3-q1)/2
    return result
```

```python
def simpangan_baku(arr):
    import numpy as np
    
    pembilang = (arr - arr.mean()) ** 2
    
    penyebut = len(arr)
    
    result = pembilang/penyebut
    result = result.sum()
    result = np.sqrt(result)
    
    return  result
```

```python
def simpangan_rata2(arr):
    import numpy as np
    
    pembilang = abs(arr - arr.mean())
    pembilang = pembilang.sum()
    
    penyebut = len(arr)
    
    result = pembilang / penyebut
    
    return result
```

```python
def ragam(arr):
    import numpy as np
    
    pembilang = (arr - arr.mean()) ** 2
    pembilang = pembilang.sum()
    
    penyebut = len(arr)
    
    result = pembilang / penyebut
    return result
```
